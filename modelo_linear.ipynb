```python
import numpy as np
import matplotlib.pyplot as plt
```

## RegressÃ£o Linear

```python
def hat_matrix(x:list)->list: 
    _denominador = np.dot(x.T,x)
    return x/_denominador
```

```python
def fit(beta:list, y:list)->list:
    return np.dot(beta,y)
```

```python
%%time
if __name__ == "__main__":
    N = 5000 #nÃºmero de pontos que quero.
    
    X = np.array([np.random.rand() for _ in range(N)]) # Gera um array com 500 ou 5000 nÃºmeros.
    noise = np.array([np.random.randn() for _ in range(N)]) # randn() fica entre 0 e 1.
    y = np.array(1.2 * X + noise)
    beta = hat_matrix(X) # matriz de valores beta.
    model = fit(beta,y)# fit ou modelo com os dados inseridos.
    y_hat = np.array(model*X)
    
    plt.scatter(X, y, alpha=0.2, label='Training Data') # alphs=0.2 ajuda a melhorar a visualizaÃ§Ã£o do grÃ¡fico, caso a visualizaÃ§Ã£o fica denso.
    plt.plot(X, y_hat, color='red', label=round(model,2))
    plt.title('Modelo de RegressÃ£o Linear - (Regression Linear Model)')
    plt.legend()
    plt.show()
    #print(y_hat)
```
